```python
# DROP BLANK COLUMNS:

df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#checking for null values

pd.isnull(df)
```

```python
pd.isnull(df).sum()

#Gives insight on presence of null values,for eg here we see amount col has 12 null values
```

```python
# drop null values

df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()    #now no null values present
```

```python
df['Amount'] = df['Amount'].astype('int')

#as the amount col had float dtype , we have to change it to int.
```

```python
df.info()  #dtype of amount has been changed successfully.
```

```python
 df.rename(columns={'Gender':'Sex'}, inplace=True)            #Renaming column
```

### EXPLORATORY DATA ANALYSIS

#### GENDER

```python
# plotting a bar chart for Gender and it's count

a = sns.countplot(x='Sex', data=df , palette="crest")

for bars in a.containers:
    a.bar_label(bars)         # gives count in graph itself.
```

```python
sales_gen = df.groupby(['Sex'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen       #to see which gender spends how much money
```

```python

sns.barplot(x='Sex',y='Amount' , data=sales_gen , palette="flare" )
```

#### From the above grahs we can infer that most of the buyers are Females and the amount spent on shopping by women is more than men.

#### AGE

```python
a = sns.countplot(data = df, x = 'Age Group', hue = 'Sex' , palette="magma")

for bars in a.containers:
    a.bar_label(bars)
```

```python
# Total Amount vs Age Group

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age , palette="viridis")
```

**From the above graphs we can infer that maximum shopping is done by Females belonging to an age group of 26-35**

#### STATE

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,7)})         #state names would overlap otherwise

sns.barplot(x='State',y='Orders', data= sales_state , palette="cubehelix")

```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,7)})         #state names would overlap otherwise

sns.barplot(x='State',y='Amount', data= sales_state , palette='YlOrBr')
```

#### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively



#### OCCUPATION

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#occupation vs Amount spent

sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_state
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

#### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

### PRODUCT CATEGORY

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Product Category vs Amount

sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

#### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category



```python
# top 10 most sold products (same thing as above)


sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

### Conclusion:
#### Women of age group 26-35 years from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
